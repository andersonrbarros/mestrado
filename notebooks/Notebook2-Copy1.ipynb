```python
import pandas as pd
```

```python
#Dataset1 (Moura, Anchieta e Marinho, 2022 - Essay-BR: a Brazilian Corpus to Automatic Essay Scoring Task)
#Redações extraídas entre Janeiro de 2015 e Agosto de 2021
```

```python
arquivo = '/home/arbarros/Mestrado/datasets/extended_essay-br_ordenada.csv'
essay_br = pd.read_csv(arquivo)
df = pd.DataFrame(essay_br)
```

```python
df
```

```python
indices_essay_vazias = df.index[(df['essay'] == "[]")].tolist()

print('Quantidade:',len(indices_essay_vazias))
print('indices das redeações vazias',indices_essay_vazias)
```

```python
indices = df.index[(df['c2'] == 0) & (df['score'] != 0)].tolist()
print(indices)
print(len(indices))
```

```python
# Eliminando as 10 redações vazias
df = df[df['essay'].notna() & (df['essay'] != "[]")]
```

```python
# Eliminando as 31 redações com inconsistencia da competência 2 ZERADA
df = df[~((df['c2'] == 0) & (df['score'] != 0))]
```

```python
df.to_csv('new_extended_essay_br.csv')
```
