```python
# Lista de possíveis notas do ENEM
possible_scores = np.arange(0, 1041, 40)  # Gera [0, 40, 80, ..., 1000]

# Garantir que cada score seja mapeado diretamente para ele mesmo se estiver na lista de possíveis valores
df_essay_br['score_class'] = df_essay_br['score'].apply(lambda x: x if x in possible_scores else None)
```

```python
df_essay_br
```

```python
from imblearn.combine import SMOTETomek

# Features (X) e target (y) para balanceamento
X = df_essay_br.drop(columns=['score', 'score_class', 'title', 'essay'])  # Excluindo colunas irrelevantes
y = df_essay_br['score_class']  # Classes discretizadas

# Aplicar SMOTE + Tomek Links
smote_tomek = SMOTETomek(random_state=42,k_neighbors=1)
X_resampled, y_resampled = smote_tomek.fit_resample(X, y)

# Verificar a distribuição das classes após balanceamento
print(y_resampled.value_counts())
```
